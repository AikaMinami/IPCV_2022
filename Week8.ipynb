```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

# **Dilation with 5x5 square**

```python
img = cv2.imread('/content/drive/MyDrive/IPCV/j.png')
kernel = np.ones((5,5), np.uint8)
dilation = cv2.morphologyEx(img, cv2.MORPH_DILATE, kernel) 

plt.subplot(121),plt.imshow(img),plt.title('First Image')
plt.subplot(122),plt.imshow(dilation),plt.title('Dilation Result')
```

# **Erotion with 5x5 square**

```python
kernel = np.ones((5,5), np.uint8)
erosion = cv2.morphologyEx(img, cv2.MORPH_ERODE, kernel)

plt.subplot(121), plt.imshow(img), plt.title('First Image')
plt.subplot(122), plt.imshow(erosion), plt.title('Erotion Result')
```

# **Opening Operation in 7x7 Square**

```python
kernel = np.ones((7,7), np.uint8)
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)

plt.subplot(121), plt.imshow(img), plt.title('First Image')
plt.subplot(122), plt.imshow(opening), plt.title('Opening Result')
```

# **Closing Operation in 7x7 square**

```python
kernel = np.ones((7,7), np.uint8)
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)

plt.subplot(121), plt.imshow(img), plt.title('First Image')
plt.subplot(122), plt.imshow(closing), plt.title('Closing Result')
```

# **Cross Morph**

## **3X3**

```python
kernel = cv2.getStructuringElement(cv2.MORPH_CROSS,(3,3)) 
dilation = cv2.dilate(img, kernel, iterations = 1) 
print("SED = Cross 3x3")
plt.subplot(121),plt.imshow(img),plt.title('First Image') 
plt.subplot(122),plt.imshow(dilation),plt.title('Cross Result')
```

## **5x5**

```python
kernel = cv2.getStructuringElement(cv2.MORPH_CROSS,(5,5)) 
dilation = cv2.dilate(img, kernel, iterations = 1) 
print("SED = Cross 5x5")
plt.subplot(121),plt.imshow(img),plt.title('First Image') 
plt.subplot(122),plt.imshow(dilation),plt.title('Cross Result')
```

## **Circular Morph**

## **3x3**

```python
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(3,3)) 
dilation = cv2.dilate(img, kernel, iterations = 1) 
print("SE = circle 3x3") 
plt.subplot(121),plt.imshow(img),plt.title('First Image') 
plt.subplot(122),plt.imshow(dilation),plt.title('Circular Result')
```

## **5x5**

```python
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(5,5)) 
dilation = cv2.dilate(img, kernel, iterations = 1) 
print("SE = circle 5x5") 
plt.subplot(121),plt.imshow(img),plt.title('First Image') 
plt.subplot(122),plt.imshow(dilation),plt.title('Circular Result')
```

# **Rect Morph**

## **3x3**

```python
kernel = cv2.getStructuringElement(cv2.MORPH_RECT,(3,3)) 
erosion = cv2.erode(img,kernel,iterations = 1) 
print("SE = Rectangle 3x3") 
plt.subplot(121), plt.imshow(img), plt.title('First Image') 
plt.subplot(122), plt.imshow(erosion), plt.title('Rect Result')
```

## **5x5**

```python
kernel = cv2.getStructuringElement(cv2.MORPH_RECT,(5,5)) 
erosion = cv2.erode(img,kernel,iterations = 1) 
print("SE = Rectangle 5x5") 
plt.subplot(121), plt.imshow(img), plt.title('First Image') 
plt.subplot(122), plt.imshow(erosion), plt.title('Rect Result')
```
