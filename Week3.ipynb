# **Linear Brightness Transformation**

Formula: *g(x,y) = f(x,y)+b*
*g(x,y)* is a value of pixels after the transformation, *f(x,y)* is the real pixel value, b is brightness value 

```python
print(" Change Vision Brightness ")
print("--------------------------")

try:
  brightness = int(input("Input brightness value: "))
except ValueError:
  print("Please input a number")

ori = cv.imread('/content/drive/MyDrive/IPCV/571.jpg')
brightness_image = np.zeros(ori.shape, ori.dtype)

for y in range(ori.shape[0]):
  for x in range(ori.shape[1]):
    for c in range(ori.shape[2]):
      brightness_image[y,x,c] = np.clip(ori[y,x,c] + brightness, 0, 255)

final_frame = cv.hconcat((ori, brightness_image))
cv2_imshow(final_frame)
```
